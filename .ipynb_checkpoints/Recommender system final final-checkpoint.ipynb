```python
TRAIN_DIR = ('data/Training/')
TEST_DIR = ('data/Testing/')
```

```python

# Load the dataset
def load_data(dir_path, IMG_SIZE):
    X, y = [], []
    labels = dict()
    i = 0
    for path in tqdm(sorted(os.listdir(dir_path))):
        if not path.startswith('.'):
            labels[i] = path
            for file in os.listdir(dir_path + path):
                if not file.startswith('.'):
                    img = cv2.imread(dir_path + path + '/' + file)
                    img = img.astype('float32') / 255
                    resized = cv2.resize(img, IMG_SIZE, interpolation=cv2.INTER_AREA)
                    X.append(resized)
                    y.append(i)
            i += 1
    X = np.array(X)
    y = np.array(y)
    print(f'{len(X)} images loaded from {dir_path} directory.')
    return X, y, labels

IMG_SIZE = (48, 48)
X_train, y_train, train_labels = load_data(TRAIN_DIR, IMG_SIZE)
X_test, y_test, test_labels = load_data(TEST_DIR, IMG_SIZE)
```

```python
import matplotlib.pyplot as plt

# Function to plot a few samples from the dataset
def plot_samples(X, y, labels_dict, num_samples=5):
    plt.figure(figsize=(10, 10))
    for class_index, class_name in labels_dict.items():
        indices = np.where(y == class_index)[0][:num_samples]
        for i, idx in enumerate(indices):
            plt.subplot(len(labels_dict), num_samples, i + 1 + (class_index * num_samples))
            plt.imshow(X[idx].reshape(48, 48, 3))  # assuming images are RGB (3 channels)
            plt.title(f'{class_name}')
            plt.axis('off')
    plt.show()

# Visualize samples from the training set
plot_samples(X_train, y_train, train_labels)

```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Mapping the numeric class indices to emotion labels
class_labels = list(train_labels.values())

# Count occurrences of each class in y_train
class_counts = Counter(y_train)

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(class_labels, [class_counts[i] for i in range(len(class_labels))], color='skyblue')

# Add labels and title
plt.xlabel('Emotion Classes')
plt.ylabel('Count')
plt.title('Distribution of Emotion Classes in the Training Set')
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

```python

# Convert labels to one-hot encoding
from tensorflow.keras.utils import to_categorical
Y_train = to_categorical(y_train, num_classes=6)
Y_test = to_categorical(y_test, num_classes=6)


def create_multinetemotion_model(IMG_SIZE):
    input_img = Input(shape=IMG_SIZE + (3,))  # Assume images are already RGB (3 channels)

    # Load pre-trained models without the top layers
    resnet50_model = ResNet50(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))
    vgg16_model = VGG16(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))

    # Flatten outputs from both models
    resnet50_output = Flatten()(resnet50_model(input_img))
    vgg16_output = Flatten()(vgg16_model(input_img))

    # Concatenate the outputs
    merged = concatenate([resnet50_output, vgg16_output])

    # Add fully connected layers
    dense1 = Dense(512, activation='relu')(merged)
    dropout1 = Dropout(0.5)(dense1)
    dense2 = Dense(256, activation='relu')(dropout1)
    dropout2 = Dropout(0.5)(dense2)
    dense3 = Dense(128, activation='relu')(dropout2)
    
    # Final output layer (for 6 emotion classes)
    output = Dense(6, activation='softmax')(dense3)

    # Create the model
    model = Model(inputs=input_img, outputs=output)
    
    return model

```

```python

# Create and compile the model
IMG_SIZE = (48, 48)
model = create_multinetemotion_model(IMG_SIZE)

# Compile the model
model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=1e-4),
    metrics=['accuracy']
)
# Summary of the model
model.summary()
```

```python
# Train the model
def deep_model(model, X_train, Y_train, X_val, Y_val, epochs, batch_size):
    history = model.fit(
        X_train,
        Y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_val, Y_val),
        verbose=1
    )
    return history

# Train the model
history = deep_model(model, X_train, Y_train, X_test, Y_test, epochs=10, batch_size=16)
```

```python

# Plot training metrics
def plot_metrics(history):
    # Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

# Evaluate and plot metrics
plot_metrics(history)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    
    plt.figure(figsize = (6,6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    cm = np.round(cm,2)
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
# Validate on test set

predictions = model.predict(X_test)
y_pred = [np.argmax(probas) for probas in predictions]


accuracy = accuracy_score(y_test, y_pred)
print('Test Accuracy = %.2f' % accuracy)

confusion_mtx = confusion_matrix(y_test, y_pred) 
cm = plot_confusion_matrix(confusion_mtx, classes = list(test_labels.items()), normalize=False)
```

```python
from sklearn.metrics import classification_report
def evaluate_model(model, X_test, y_test, test_labels):
    # Evaluate on test set
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    print('Test Loss:', loss)
    print('Test Accuracy:', accuracy)

    # Confusion matrix
    predictions = model.predict(X_test)
    y_pred = np.argmax(predictions, axis=1)
    confusion_mtx = confusion_matrix(np.argmax(y_test, axis=1), y_pred)
    plot_confusion_matrix(confusion_mtx, classes=list(test_labels.values()), normalize=False)

    # Classification report
    target_names = list(test_labels.values())
    print(classification_report(np.argmax(y_test, axis=1), y_pred, target_names=target_names))

evaluate_model(model, X_test, Y_test, test_labels)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess the images from the directory
def load_sample_images(dir_path, img_size, num_samples=20):
    images = []
    true_labels = []
    i = 0
    for class_dir in sorted(os.listdir(dir_path)):
        if not class_dir.startswith('.'):
            class_path = os.path.join(dir_path, class_dir)
            for file in os.listdir(class_path):
                if i >= num_samples:
                    break
                img_path = os.path.join(class_path, file)
                img = cv2.imread(img_path)
                img = cv2.resize(img, img_size)
                img = img.astype('float32') / 255
                images.append(img)
                true_labels.append(class_dir)  # Store the true class name
                i += 1
        if i >= num_samples:
            break
    return np.array(images), true_labels

# Convert grayscale images to 3-channel images
def convert_to_rgb(images):
    rgb_images = []
    for img in images:
        if len(img.shape) == 2 or img.shape[-1] == 1:  # If grayscale, convert to RGB
            img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        else:
            img_rgb = img  # If already RGB, no conversion
        rgb_images.append(img_rgb)
    return np.array(rgb_images)

# Predict and display images along with actual and predicted labels
def display_predictions(model, images, true_labels, class_labels):
    plt.figure(figsize=(12, 12))
    for i in range(len(images)):
        # Prepare the image for prediction
        img = np.expand_dims(images[i], axis=0)  # Add batch dimension
        
        # Make prediction
        predictions = model.predict(img)
        predicted_class = class_labels[np.argmax(predictions)]
        
        # Display the image and labels
        plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].astype('float32'))  # Display the image
        plt.title(f"True: {true_labels[i]}\nPred: {predicted_class}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Example usage
IMG_SIZE = (48, 48)
class_labels = ['Angry', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# Load 20 samples from the testing directory
X_samples, y_true = load_sample_images(TEST_DIR, IMG_SIZE, num_samples=20)

# Convert grayscale images to RGB for the model
X_samples_rgb = convert_to_rgb(X_samples)

# Display the predictions with actual and predicted classes
display_predictions(model, X_samples_rgb, y_true, class_labels)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

class_labels = ['Angry', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
# Function to preprocess the new image (colored image)
def preprocess_new_image(image_path, img_size=(48, 48)):
    # Load the image in color
    img = cv2.imread(image_path)

    # Resize the image to the required size (48x48)
    img_resized = cv2.resize(img, img_size)

    # Normalize the pixel values (0 to 1)
    img_resized = img_resized.astype('float32') / 255.0

    # Reshape to match the input shape for the model (1, 48, 48, 3)
    img_resized = np.expand_dims(img_resized, axis=0)   # Add the batch dimension

    return img_resized

# Predict the emotion of the new image
def predict_new_image(image_path, model, class_labels, img_size=(48, 48)):
    # Preprocess the image
    preprocessed_img = preprocess_new_image(image_path, img_size)

    # Predict the emotion
    predictions = model.predict(preprocessed_img)

    # Get the predicted class
    predicted_class = class_labels[np.argmax(predictions)]

    return predicted_class

# Example usage
image_path = 'image.jpg'  # Path to the new image
predicted_emotion = predict_new_image(image_path, model, class_labels)

# Display the prediction
print(f"The predicted emotion is: {predicted_emotion}")

# Display the image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying
plt.imshow(img_rgb)
plt.title(f"Predicted Emotion: {predicted_emotion}")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np

# Function to preprocess the frame for prediction
def preprocess_frame(frame, img_size=(48, 48)):
    # Convert to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Resize to match model input
    resized_frame = cv2.resize(gray_frame, img_size)
    # Normalize pixel values
    normalized_frame = resized_frame.astype('float32') / 255
    # Convert to 3 channels (repeat the grayscale values)
    three_channel_frame = np.stack((normalized_frame,) * 3, axis=-1)
    # Add batch dimension
    three_channel_frame = np.expand_dims(three_channel_frame, axis=0)
    return three_channel_frame

# Start video capture
cap = cv2.VideoCapture(0)  # 0 for default camera

while True:
    ret, frame = cap.read()  # Read a frame from the camera
    if not ret:
        break  # Break the loop if no frame is captured

    preprocessed_frame = preprocess_frame(frame)  # Preprocess the frame

    # Make prediction
    predictions = model.predict(preprocessed_frame)
    predicted_class = class_labels[np.argmax(predictions)]

    # Display the prediction on the frame
    cv2.putText(frame, f'Emotion: {predicted_class}', (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    
    # Show the video feed with the prediction
    cv2.imshow('Emotion Detection', frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import time

# Define your emotion labels and the corresponding recommendations
class_labels = ['Angry', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
recommendations = {
    'Happy': 'Continue doing what you are doing!',
    'Neutral': 'Continue with your activities.',
    'Sad': 'We recommend watching a comic or cartoon video.',
    'Angry': 'How about watching a comic or cartoon video?',
    'Fear': 'Consider watching something lighthearted, like a comic or cartoon.',
    'Surprise': 'That’s interesting! Keep it up!'
}

# Function to preprocess the frame from the camera
def preprocess_frame(frame):
    img = cv2.resize(frame, (48, 48))  # Resize to match model input
    img = img.astype('float32') / 255  # Normalize
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img

# Open the camera
cap = cv2.VideoCapture(0)

# Start the timer
start_time = time.time()
duration = 10 * 60  # 10 minutes in seconds
emotion_counts = {label: 0 for label in class_labels}  # Initialize emotion counts

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    preprocessed_frame = preprocess_frame(frame)  # Preprocess the frame

    # Make prediction
    predictions = model.predict(preprocessed_frame)
    predicted_class = class_labels[np.argmax(predictions)]

    # Count the detected emotion
    emotion_counts[predicted_class] += 1

    # Display the prediction on the frame
    cv2.putText(frame, f'Emotion: {predicted_class}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.imshow('Emotion Detection', frame)

    # Check if 10 minutes have passed
    if time.time() - start_time > duration:
        # Determine the overall mood based on detected emotions
        happy_or_neutral = emotion_counts['Happy'] + emotion_counts['Neutral']
        sad_angry_fear = emotion_counts['Sad'] + emotion_counts['Angry'] + emotion_counts['Fear']

        if happy_or_neutral > sad_angry_fear:
            print(recommendations['Happy'])  # Can be replaced with a GUI pop-up or similar
        else:
            print(recommendations['Sad'])  # Can be replaced with a GUI pop-up or similar

        break

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()

```
